```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data
nnfs.init()




class DenseLayer:
    def __init__(self, n_inputs, n_neurons):
        self.weights= 0.01*np.random.randn(n_inputs,n_neurons)
        self.biases= np.zeros((1,n_neurons))
    
    def forward(self,inputs):
        self.output=np.dot(inputs,self.weights) + self.biases
        return self.output
    
    def forward_activation(self,inputs):
        self.active_relu= np.maximum(0,inputs)
        return self.active_relu

X,y= spiral_data(samples=100, classes=3)


layer1=DenseLayer(2,3)
lf1=layer1.forward(X)
l1_final=layer1.forward_activation(lf1)

layer2= DenseLayer(3,2)
lf2= layer2.forward(lf1)
l2_final= layer2.forward_activation(lf2)

print(l2_final[:5])





# layer2= DenseLayer(3,2)
# layer2.forward(lf1)

# print(layer2.output[:5])

        
```

Softmax

```python
inputs= [[1,2,3,2.5],
         [2., 5.,-1., 2],
         [-1.5, 2.7,3.3,-0.8]]

exp_values= np.exp(inputs- np.max(inputs, axis=1, keepdims=True))

probabilities= exp_values/ np.sum(exp_values, axis=1, keepdims=True)
print(probabilities)
print(np.sum(probabilities, axis=1))
```

Implementing Softmax for the classification problem

```python
inputs= [[1,2,3,2.5],
         [2., 5.,-1., 2],
         [-1.5, 2.7,3.3,-0.8]]

class Softmax:

    def activation_softmax(self,inputs):
        exp_values= np.exp(inputs- np.max(inputs, axis=1, keepdims=True))
        probabilities=  exp_values/np.sum(exp_values, axis=1, keepdims=True)
        return probabilities

sf_act=Softmax()
sf_act.activation_softmax(inputs)
```
